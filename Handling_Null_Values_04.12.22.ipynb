```python
df7 = df.fillna(method = 'bfill', axis =1) # fill it with right value
df7
```

```python
# fill it with different columns.
df8 = df.fillna({'society': 'abcd',
                'balcony': 'defg'})
df8
```

```python
# filling null value with the mean of the column

df9 = df.fillna(value = df['balcony'].mean())
df9
```

```python
df9 = df.fillna(value = df['balcony'].max())
df9
```

```python
# dropna()
df10 = df.dropna()
df10
```

```python
df11 = df.dropna(how='all')
df11              #how takes 2 value any.all
```

```python
#. replace()
import numpy as np
df12 = df.replace(to_replace = np.nan, value=875465)
df12
```

```python
df13 = df.replace(to_replace = 3.0, value = 5.0)
df13
```

```python
# interpolate()

df['balcony'] = df['balcony'].interpolate(method = 'linear')
df
```

```python
df['balcony'] = df['balcony'].interpolate(method = 'linear', limit_direction='forward')
df 
```
