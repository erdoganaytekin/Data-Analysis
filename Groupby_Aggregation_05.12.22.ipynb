```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('planets')
df.head()
```

```python
df.groupby('method')[['distance']].sum()
```

```python
df.groupby('method').sum()
```

```python
df.groupby('method').mean()
```

```python
df.groupby('orbital_period').aggregate([np.min,np.max,np.median])
```
